```python
#Question 1 : Derivative by forward method. (sin function)

import numpy as np
import matplotlib.pyplot as plt

def g(x):
    y = np.sin(x)
    return y

h = 0.1

def f(x):
    y = (g(x+h)-g(x))/h
    return y

n = float(input("Enter angle value in radians :"))
k = f(n)
print("The derivative value is : ",k)

xaxis = np.arange(-2*np.pi,2*np.pi,0.001)
yaxis = f(xaxis)

x1 = np.arange(-2*np.pi,2*np.pi,0.001)
y1 = np.cos(x1)

plt.plot(x1,y1,label ="Mathematical Derivative")

plt.plot(xaxis,yaxis,label="Analytical Derivative")
plt.legend()
plt.show()
```

```python
#Question 1 : Derivative by forward method. (cosine function)

import numpy as np
import matplotlib.pyplot as plt

def g(x):
    y = np.cos(x)
    return y

h = 0.1

def f(x):
    y = (g(x+h)-g(x))/h
    return y

n = float(input("Enter angle value in radians :"))
k = f(n)
print("The derivative value is : ",k)

xaxis = np.arange(-2*np.pi,2*np.pi,0.001)
yaxis = []

i = -2*np.pi
while i >= -2*np.pi and i <= 2*np.pi:
    r = f(i)
    yaxis.append(r)
    i = i + 0.001

x1 = np.arange(-2*np.pi,2*np.pi,0.001)
y1 = -np.sin(x1)

plt.plot(x1,y1,label ="Mathematical Derivative")

plt.plot(xaxis,yaxis,label="Analytical Derivative")
plt.legend()
plt.show()
```
