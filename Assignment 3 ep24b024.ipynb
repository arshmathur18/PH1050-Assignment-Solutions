```python
#Question 2 : Plot Sin(x)+Cos(y)*Sin(y)*Cos(x)*Cos(x) for x,y range −2π to 2π. Use a 2D grid with dimensions
# 128 × 128.

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

x = np.linspace(-2*np.pi, 2*np.pi,128)
y = np.linspace(-2*np.pi, 2*np.pi,128)
X,Y = np.meshgrid(x,y) #meshgrid aranges x and y on a coordinate plane
z = np.sin(X) + np.cos(Y)*np.sin(Y)*np.cos(X)*np.cos(X)

plt.pcolormesh(X,Y,z,cmap='magma')
plt.colorbar()

plt.show()
```

```python
#question 3 : Plot a 2D Gaussian function for x = −5 to +5 and y = −5 to +5.

import matplotlib.pyplot as plt 
import numpy as np

fig = plt.figure()

x = np.linspace(-5, 5, 1000) 
y = x.copy()
z = np.exp(-(x ** 2 + y ** 2))

plt.plot(y,z)
plt.show()

```
