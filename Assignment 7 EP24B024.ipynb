```python
#newton-raphsen method practice
import numpy as np
import matplotlib.pyplot as plt

def fn(x):
    return x**2 - 4

x1=np.arange(-10,10,0.001)
y=fn(x1)
plt.plot(x1,y)
plt.grid()
plt.show()

def diff(f,a):
    h = 0.01
    q=(f(a+h)-f(a))/h
    return q

def nr_method(f,guess):
    for i in range(1000):
        root = guess - f(guess)/diff(f,guess)
        guess = root #recursion

    return root

print(nr_method(fn,-1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fn(x):
    return x**4 - 4

x1=np.arange(-2,2,0.001)
y=fn(x1)
plt.plot(x1,y)
plt.grid()
plt.show()

def diff(f,a):
    h = 0.01
    q=(f(a+h)-f(a))/h
    return q

def nr_method(f, guesses, tol=1e-8, max_iter=1000):
    roots = []

    for guess in guesses:
        for i in range(max_iter):
            next_guess = guess - f(guess) / diff(f, guess)
            
            if abs(next_guess - guess) < tol:
                root = next_guess
                break 
            guess = next_guess
        else:
            continue
        if not any(abs(root - r) < tol for r in roots):
            roots.append(root)

    return roots

print(nr_method(fn,[-5,5]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random

def f(x):
    return x**2 - 4

x_axis = np.linspace(-4,4,100)
y_axis = f(x_axis)

plt.figure()
plt.grid()
plt.plot(x_axis,y_axis)
plt.show()

def bisection(f,step):
    x0 = float(input('enter lower limit'))
    xend = float(input('enter upper limit'))
    x_data = np.arange(x0,xend,step)
    a = x0
    b = x0 + step
    roots = []
    while b <= x_data[-1]:
        
        if f(a)*f(b) < 0:
            x = (a+b)/2
            
            if f(a)*f(x) < 0:
                b = x
            elif f(b)*f(x) < 0 :
                a = x

            if abs(a-x) < 0.000001:
                roots.append(x)
            
            if abs(b-x) < 0.000001:
                roots.append(x)

            a += step
            b += step

        else :
            a += step
            b += step
            
    return roots

print(bisection(f,0.001)) #step size should be greater than tolerance for this to work 

def derivative(x):
    h = 0.001
    y = (f(x+h) - f(x-h))/(2*h)
    return y

def newtonraphsen(f,step):
    
    x0 = float(input('enter lower limit'))
    xend = float(input('enter upper limit'))
    x_data = np.arange(x0,xend,step)
    roots = []
    for j in range(100):
    
        x = x0 + random.rand()*(xend - x0)
    
        for i in range(len(x_data)):
            
            x1 = x - f(x)/derivative(x) 
           
            if abs(x-x1) < 0.000001 :
                root = x1
                break
                
            x = x1
        else:
            continue
            
        if not any(abs(root - r) < 0.000001 for r in roots):
            roots.append(root)

    return roots

print(newtonraphsen(f,0.0001))
```
