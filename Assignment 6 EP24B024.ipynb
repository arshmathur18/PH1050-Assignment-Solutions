```python
import numpy as np

def trapezium(f,a,b,N):
    h = (b-a)/N
    I = []
    
    I.append(f(a)/2)
    I.append(f(b)/2)
    i = 1
    for i in range(N-1):
        I.append(f(a+i*h))

    I1 = np.array(I)
    return h*np.sum(I1)

def f(x): 
    return x**2 

print(trapezium(f,2,6,1000))

def simpsons(f,a,b,N):  
    h = (b-a)/N
    I = []
    I.append(f(a))
    I.append(f(b))
    i = 1
    while i < N/2 :
        I.append(4*f(a + h*(2*i - 1)) + 2*f(a + 2*i*h)) 
        i = i + 1
    I1 = np.array(I)
    return (h/3)*np.sum(I1)

print(simpsons(f,2,6,1000))
```
